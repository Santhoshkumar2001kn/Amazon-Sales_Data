```python
# Histogram

df['Cate'].hist()
```

```python
df['Category'] = df['Category'].astype(str)
column_data = df['Category']
plt.figure(figsize=(10, 5))
plt.hist(column_data, bins=30, edgecolor='Black')
plt.xticks(rotation=90)
plt.show()
```

```python
# Checking B2B Data  by using pie chart 
Status_Check = df['Status'].value_counts()

#  Plot the pie chart
plt.pie(Status_Check, labels=Status_Check, autopct='%1.1f%%')
#plt.axis('equal')
plt.show()
```

```python
# Checking B2B Data  by using pie chart 
Status_Check = df['Status'].value_counts()

#  Plot the pie chart
plt.pie(Status_Check, labels=Status_Check.index, autopct='%1.1f%%')
#plt.axis('equal')
plt.show()
```

```python
#  Prepare data for pie chart
a1 = df['Payment Method'].value_counts()

# Step 4: Plot the pie chart
fig, ax = plt.subplots()

ax.pie(a1, labels=a1.index, autopct='%1.1f%%', radius=0.7, wedgeprops=dict(width=0.6))
ax.set(aspect="equal")

plt.show()
```

```python
# Prepare data for scatter plot
plt.figure(figsize=(10,5))

x_data = df['Customer Location']  
y_data = df['Date'] 

# Plot the scatter plot
plt.scatter(x_data, y_data)
plt.xlabel('Customer Location')  
plt.ylabel('Date')  
plt.title('Scatter Plot') 
plt.show()
```

```python
# Plot count of cities by location
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Customer Location')
plt.xlabel('Location')
plt.ylabel('count')
plt.title('Distribution of Places')
plt.xticks(rotation=90)
plt.show()
```

```python
# top_5_Location 
top_5_loc = df['Customer Location'].value_counts().head(5)
# Plot count of cities by location
sns.countplot(data=df[df['Customer Location'].isin(top_5_loc.index)], x='Customer Location')
plt.xlabel('Location')
plt.ylabel('count')
plt.title('Distribution of Places')
plt.xticks(rotation=45)
plt.show()
```
